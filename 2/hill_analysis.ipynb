CRYPTANALYSIS OF HILL CIPHER


Hill cipher is one of the encryption techiniques in which a matrix of plain text is multiplied by a square key 
matrix in which the keys have inverse in modulo arithmetic and the key degree and the column numbers of 
plain text is decided as same.

Some of the cryptanalysis methods used are :-

METHOD 1 : CYPHER TEXT ONLY ATTACK

This is where the adversary has only one or a few cypher texts with her and then since the algorithms are made public
she does analysis on the cypher text to find the key.


Brute Force Attak : Here all the possible keys are tried out . And also we have to try it for different column lengths 
                    as well . This would be an awfully large amount of outputs to filter from .
